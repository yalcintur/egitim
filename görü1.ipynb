```python
path = "/content/dataset"
dls = ImageDataLoaders.from_folder(path, valid_pct=0.2, item_tfms=Resize(224), bs=32)
```

```python
dls.show_batch()
```

```python
import cv2 as cv
from google.colab.patches import cv2_imshow

def predict(path_to_img, learner):
  img = cv.imread(path_to_img)
  cv2_imshow(img)
  print(learner.predict(path_to_img))
```

```python
predict("/content/dataset/kedi/cat.0.jpg", learn)
```
